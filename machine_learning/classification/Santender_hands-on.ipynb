```python
# var3값 이상하게 들어간 값 대체하기
cust_df['var3'].value_counts()[:10]
```

```python
cust_df['var3'].replace(-999999, 2, inplace=True)
# 필요없는 feature인 ID 삭제
cust_df.drop('ID', axis=1, inplace=True)
```

```python
# X, Y 데이터프레임으로 각각 분류하기
x_features = cust_df.iloc[:, :-1]
y_labels = cust_df.iloc[:, -1]
print("features 데이터 shape", x_features.shape)
```

```python
# train_test split하고 난 다음 class imbalance여부 관찰하기
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_features,
                                                   y_labels,
                                                   test_size=0.2,
                                                   random_state=42,
                                                   stratify=y_labels)

train_cnt = y_train.count()
test_cnt = y_test.count()

print("train 데이터의 label 분포\n", y_train.value_counts()/train_cnt)
print("test 데이터의 label 분포\n", y_test.value_counts()/test_cnt)
```

```python
# 위에서 분할한 데이터로 XGBoost 모델링하기
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score

# 모델 구축
xgb_clf = XGBClassifier(n_estimators=300, random_state=42)
# 모델 학습(검증 데이터를 test데이터로 사용 - 실제 이러면 안 됨.)
xgb_clf.fit(x_train, y_train, early_stopping_rounds=80,
           eval_metric='auc',
           eval_set=[(x_test, y_test)])
# average ='macro' 는 class imbalance를 고려하지 않음
xgb_roc_score = roc_auc_score(y_test,
                              xgb_clf.predict_proba(x_test)[:,1],
                             average='macro')
print(f"ROC AUC : {xgb_roc_score: .4f}")
```
