```python
import requests
from bs4 import BeautifulSoup
import pymysql
db = pymysql.connect(host='localhost', port=3306, user='root', passwd='watson1259@', db='yes24_best',charset='utf8')
cursor = db.cursor()

res = requests.get("http://www.yes24.com/24/category/bestseller")
soup = BeautifulSoup(res.content, 'html.parser')

categories = soup.select("ul.caTab_2nd li")
for category in categories:
    name = category.select_one('a').get_text()
    link = category.select_one('a')['href']
    multi_requests(name, f"http://www.yes24.com/{link}")

db.commit()
db.close()
```

```python
def multi_requests(category_name, url):
    print(f"Starting '{category_name}' category page")
    get_data(category_name, url)
```

```python
def get_data(category_name, url):
    res = requests.get(url)
    soup = BeautifulSoup(res.content, 'html.parser')
    results = soup.select("div#bestList ol li")
    for result in results:
        data_dict = dict()
        title = result.select_one("p:nth-of-type(3) a")
        if title != None:
            title = title.get_text()
        author = result.select_one("p:nth-of-type(4) > a")
        if author != None:
            author = author.get_text()
        publisher = result.select_one("p:nth-of-type(4) > a:nth-of-type(2)")
        if publisher != None:
            publisher = publisher.get_text()
        else:
            publisher = '출판사 없음'
        price = result.select_one("p.price > strong")
        if price != None:
            price = price.get_text().replace("원",'').replace(',','')
        review_num = result.select_one("p:nth-of-type(6) > a")
        if review_num != None:
            if review_num.get_text() == '':
                review_num = 0
            else:
                review_num = review_num.get_text().replace('개','')
        else:
            review_num = 0 
        book_code = result.select_one("p.copy > a")
        if book_code != None:
            book_code = book_code['href']
            book_code = book_code.split("Goods/")[1]
            print({'분류':category_name, '제목':title, '작가':author, '출판사':publisher, '가격': price,'리뷰개수': review_num,
                   '책 코드':book_code})
            data_dict['분류'] = category_name
            data_dict['제목'] = title
            data_dict['작가'] = author
            data_dict['출판사'] = publisher
            data_dict['가격'] = price
            data_dict['리뷰개수'] = review_num
            data_dict['책 코드'] = book_code
            save_data(data_dict)

        
    
```

```python
import pymysql
db = pymysql.connect(host='localhost', port=3306, user='root', passwd='watson1259@', db='yes24_best',charset='utf8')
cursor = db.cursor()

sql = """
    CREATE TABLE products(
        book_code INT NOT NULL PRIMARY KEY,
        author VARCHAR(50) NOT NULL,
        publisher VARCHAR(60) NOT NULL,
        price INT NOT NULL,
        review_num int NOT NULL
);
"""
cursor.execute(sql)

sql = """
    CREATE TABLE category(
        num INT NOT NULL AUTO_INCREMENT PRIMARY KEY,
        name VARCHAR(30) NOT NULL,
        book_code INT NOT NULL,
        title VARCHAR(200) NOT NULL,
        FOREIGN KEY(book_code) REFERENCES products(book_code)
);
"""
cursor.execute(sql)

db.commit()
db.close()
```
