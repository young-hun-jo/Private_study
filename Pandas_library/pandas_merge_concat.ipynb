```python
# merge는 기본적으로 공통 칼럼을 가진 값들중 공통의 값들 기준으로 합침
pd.merge(df1, df2)
```

```python
# merge의 다양한 방법
# 1.innder join
pd.merge(df1, df2, how='inner')
```

```python
#2. outer join
pd.merge(df1, df2, how='outer')
```

```python
#3. left outer join
pd.merge(df1, df2, how='left')
```

```python
#4. right outer join
pd.merge(df1, df2, how='right')
```
