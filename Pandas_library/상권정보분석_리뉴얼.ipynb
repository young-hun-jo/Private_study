### matplotlib의 plot은 unstack 상태로 만들어주어야만 그래프가 이쁘게 표시됨
 - plot은 index값을 보통 x축으로 그려줌(barh()는 y축에 index넣어줌) 
 
### seaborn은 melt를 써서 칼럼에 있는 값들이 모두 행으로 오도록 해주어야 함

### barplot은 subplot그리기가 안되기 때문에 catplot 사용(ci디폴트는 95% 신뢰구간에서 randomly sampling), 그래프종류(kind)도 설정 가능, sharex=False 하면 서브플롯의 x축에 모두 label새겨짐

### 특정 칼럼명 데이터 갖고오기
- 하나만 갖고올 때 : df['칼럼명'] == '칼럼1' 
- 두개 이상 갖고올 때(isin 사용) : df['칼러명'].isin(['칼럼1','칼럼2'...])

### df[df['칼럼명'] == '칼럼1'] 과 df.loc[df['칼럼명'] =='칼럼'] 의 결과값은 동일

```python
df_ipsy = df.loc[df['상권업종소분류명']=='학원-입시']
df_ipsy.shape
```

```python
import folium
```

```python
df_ipsy.head(1)
```

```python
lat = df_ipsy['위도'].mean()
long = df_ipsy['경도'].mean()

map = folium.Map(location=[lat,long], zoom_start=13)

for n in df_ipsy.index[:10]:
    lat = df_ipsy.loc[n, '위도']
    long = df_ipsy.loc[n, '경도']
    tool_tip = df_ipsy.loc[n, '상호명']+"-"+df_ipsy.loc[n, '도로명']
    
    folium.CircleMarker(location=[lat, long], tooltip=tool_tip, radius=4,
                      color='red').add_to(map)

#한글인코딩깨짐
#map.save("encoding.html")
map
```
