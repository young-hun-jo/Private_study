```python
df_academy = data[(data['시도명'] == '대전광역시') & (data['상권업종대분류명'] == '학문/교육')].copy()
df_academy.head()
```

```python
academy_count = df_academy['상권업종소분류명'].value_counts().head(30)
academy_count
```

```python
academy_count_150 = academy_count[academy_count > 150]
academy_count_150
```

```python
gu_academy = df_academy.groupby(['시군구명','상권업종소분류명'])['상호명'].value_counts()
gu_academy
```

```python
gu_academy = df_academy.groupby(['시군구명','상권업종소분류명'])['상호명'].count()
gu_academy = gu_academy.reset_index()
gu_academy = gu_academy.rename(columns={'상호명':'상호수'})
gu_academy
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(7,5))
plt.xlabel('시군구명', fontsize=15)
plt.ylabel('상호수', fontsize=15)
sns.barplot(data=gu_academy, x='시군구명', y='상호수')
```
