```python
def get_string(text):
    # nan이 들어갔을 때는 그냥 nan으로 반환
    # pd.isna(텍스트) -> T/F로 반환
    if pd.isna(text):
        return text
    
    text = re.sub("[\d\.'\-(nan)\[\]\/\+ABDEmㄱN]","",text)
    text = re.sub(" ","",text)
    text = re.sub(",{2,}",",",text)
    return text
```

```python
# "화장실,주차장," 으로 만들기
get_string("[1. '화장실' 5, 2. 주차장, -, nan]")
```

```python
# map 함수 : 리스트의 요소를 지정된 함수로 처리해줌(map함수는 원본리스트를 건드리지 않고 새로운 리스트를 만들어냄!)

# 새로운 칼럼(파생변수)만들기
df['편익시설'] = df['공원보유시설(편익시설)'].map(get_string)
df['편익시설']
```

```python
df['편익시설'].value_counts().tail()
```

```python
# join 함수 : 특정 구분자를 포함하여 문자열로 변환해줌!
# nan 값은 문자열로 변환못해줌! 그래서 notnull()로 nan값이 아닌 것들만 변환

fac = ",".join(df.loc[df['편익시설'].notnull(), '편익시설'])
fac = re.sub(",{2,}",",",fac)
fac
```

```python
phone = "+82-10-1234-5678"
phone_2 = "+82-10-134-5678"

def phone_number(phone):
    num = phone.split("-")
    num_len = len(num) // 2
    num[num_len] = re.sub("\w","*",num[num_len])
    num[-1] = re.sub("(\d{2})(\d{2})",r"**\2",num[-1])
    return "-".join(num)

print(phone_number(phone))
print()
print(phone_number(phone_2))
```
