```python
import pandas as pd
import numpy as np
```

```python
url = 'https://production-api.coinbit.global/api/v1.0/trading_pairs/'
res = requests.get(url)
soup = bs(res.content, 'html.parser')
bitcoin_dict = json.loads(str(soup))
bitcoin_code = pd.DataFrame(bitcoin_dict)
coin_code = bitcoin_code['name'].tolist()
coin_code

#not_columns = ['candle_acc_trade_value','candle_acc_trade_volume','prev_close_price','signed_change_price','signed_change_rate']
coin_info = []
for code in coin_code:
    sub_url = 'https://production-api.coinbit.global/api/v1.0/day_candles/?trading_pair='
    res = requests.get(sub_url+code)
    soup = bs(res.content, 'html.parser')
    price_dict = json.loads(str(soup))
    price_results = price_dict.get('results')
    price_df = pd.DataFrame(price_results)
        
    coin_name.append(price_df)
    
```

```python
coin_info
```

```python
data_list = []
for df in coin_info:
    data_list.append(df)
appended_data = pd.concat(data_list)
```

```python
appended_data
```

```python
final_data = appended_data.reset_index(drop=True)
final_data
```

```python
# 필요없는 칼럼 제거
not_columns = ['candle_acc_trade_value',
               'candle_acc_trade_volume',
               'prev_close_price','signed_change_price','signed_change_rate']

final_data = final_data.drop(not_columns, axis=1)
final_data.head()
```

```python
# 칼럼 값들 float형으로 수정
final_data[['open_price','high_price','low_price','close_price']] = final_data[['open_price','high_price','low_price','close_price']].astype(float)
final_data.dtypes
```

```python
# 소수점 첫째짜리만 출력
final_data[['open_price','high_price','low_price','close_price']] = final_data[['open_price','high_price','low_price','close_price']].round(0)
final_data.head()
```

```python
# date_time 칼럼 datetime으로 바꿔주기
final_data['candle_start_date_time'] = final_data['candle_start_date_time'].str.split('T', expand=True)[0]
final_data['candle_start_date_time'] = pd.to_datetime(final_data['candle_start_date_time'])
final_data.dtypes
```
