```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time

chrome_driver = "C:/webdriver/chrome/chromedriver.exe"
driver = webdriver.Chrome(chrome_driver)
driver.get("https://www.python.org/")
assert "Python" in driver.title

element = driver.find_element_by_name('q')
element.send_keys('Python')
element.send_keys(Keys.RETURN) #리턴은 대문자로!
assert "No results found." not in driver.page_source

time.sleep(2)

h3s = driver.find_elements_by_tag_name('h3')
for h3 in h3s:
    print(h3.text)
    
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import time

chromedriver = "C:/webdriver/chrome/chromedriver.exe"
driver = webdriver.Chrome(chromedriver)
driver.get("https://news.v.daum.net/v/20200207142917726")

loop, count = True, 0
while loop and count < 10:
    try:
        element = WebDriverWait(driver, 5).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "#alex-area > div > div > div > div.cmt_box > div.alex_more > a"))
        )
        more_button = driver.find_element_by_css_selector("#alex-area > div > div > div > div.cmt_box > div.alex_more > a")
        webdriver.ActionChains(driver).click(more_button).perform()
        count += 1
        time.sleep(2)
    except TimeoutException:
        loop = False
        
whole_comment= driver.find_element_by_css_selector("#alex-area > div > div > div > div.cmt_box > ul.list_comment")
comments = whole_comment.find_elements_by_tag_name("li")
for num, comment in enumerate(comments):
    comment = comment.find_element_by_css_selector("div > p").text
    print(f"{num+1} :", comment.strip().strip("\n"))
    print()
    


driver.quit()
        
```
