```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
import numpy as np
import pandas as pd
```

```python
# 칼럼의 값들 담을 리스트 할당
title_rslt = []
press_rslt = []
review_rslt = []
section_rslt = []
# 반복문 돌려서 섹션-> 페이지로 재귀적 크롤링
for section in range(100,106):
    for num in range(1, 16):
        if num >0 and num < 10:
            url = f'https://news.naver.com/main/ranking/popularDay.nhn?rankingType=popular_day&sectionId={section}&date=2020050{num}'
        else:
            url = f'https://news.naver.com/main/ranking/popularDay.nhn?rankingType=popular_day&sectionId={section}&date=202005{num}'
        res = requests.get(url)
        soup = bs(res.content, 'html.parser')
        ranks = soup.select('ol.ranking_list')
        for rank in ranks:
            contents = rank.select('div.ranking_text')
        
            for i, content in enumerate(contents):
                # 기사제목 갖고오기
                title= content.select_one('div.ranking_headline a')['title']
                # 언론사 갖고오기
                press = content.select_one('div.ranking_office').get_text()
                # 기사 본 횟수 가져와서 int형으로 바꿔주기
                review = content.select_one('div.ranking_view').get_text()
                review = review.replace(',','')
                review = int(review)
                
                # 칼럼에 값들 append하기
                title_rslt.append(title)
                press_rslt.append(press)
                review_rslt.append(review)
                section_rslt.append(section)
```

```python
# 데이터프레임 만들어 주기 위해서 빈 dict 할당
dicted_data = dict()

# 칼럼 제목(key)에 해당 칼럼에 맞는 값들(value) 할당
dicted_data['기사제목'] = title_rslt
dicted_data['언론사'] = press_rslt
dicted_data['리뷰횟수'] = review_rslt
dicted_data['섹션'] = section_rslt

# 데이터 프레임 만들어주기
df = pd.DataFrame(dicted_data)
# 만들은 데이터프레임 미리보기
df.head()
```

```python
# 섹션별로 기사 개수 확인
df['섹션'].value_counts()
```

```python
# 섹션 값(value) type 확인
df['섹션'].dtypes
```

```python
# 섹션별 id에 맞게 섹션 이름으로 바꾸어주기
df.loc[df['섹션'] == 100, '섹션'] = '정치'
df.loc[df['섹션'] == 101, '섹션'] = '경제'
df.loc[df['섹션'] == 102, '섹션'] = '사회'
df.loc[df['섹션'] == 103, '섹션'] = '생활/문화'
df.loc[df['섹션'] == 104, '섹션'] = '세계'
df.loc[df['섹션'] == 105, '섹션'] = 'IT/과학'
```
