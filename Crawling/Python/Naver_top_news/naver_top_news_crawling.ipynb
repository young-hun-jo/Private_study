```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
import numpy as np
import pandas as pd
```

```python
res = requests.get("https://news.naver.com/main/ranking/popularDay.nhn?rankingType=popular_day&sectionId=100&date=20200515")
soup = bs(res.content, 'html.parser')
soup
```

```python
# 한 페이지의 뉴스제목만 가져오기
dict_data = dict()
ranks = soup.select('ol.ranking_list')
for rank in ranks:
    contents = rank.select('div.ranking_text')
    title_rslt = list()
    press_rslt = list()
    review_rslt = list()
    for i, content in enumerate(contents):
        # 기사제목 갖고오기
        title= content.select_one('div.ranking_headline a')['title']
        # 언론사 갖고오기
        press = content.select_one('div.ranking_office').get_text()
        # 기사 본 횟수 가져와서 int형으로 바꿔주기
        review = content.select_one('div.ranking_view').get_text()
        review = review.replace(',','')
        review = int(review)
        title_rslt.append(title)
        press_rslt.append(press)
        review_rslt.append(review)
    dict_data['기사제목'] = title_rslt
    dict_data['언론사'] = press_rslt
    dict_data['리뷰횟수'] = review_rslt
    print(dict_data)
        
```

```python
df = pd.DataFrame(dict_data)
df.head()
```
