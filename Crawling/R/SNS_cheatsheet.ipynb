```r
pcntEncoding<-function(x){
    library(urltools)
    if (localeToCharset() == 'UTF-8'){
        x<-x %>% icnov(from='UTF-8', to='CP949')}
    else if (localeToCharset() %in% c('CP949','EUC-KR')){
        x<-x
        }
    string<- x %>% url_encode() %>% I()
    return(string)
}
```

```r
library(httr)
library(urltools)
library(rvest)
library(rjson)
library(tidyverse)
# 코스피 200 주식 이름,가격
URL='https://finance.naver.com/sise/entryJongmok.nhn'
name<-GET(url=URL) %>%
  read_html() %>%
  html_nodes(css='td.ctg a') %>%
  html_text()
# 주식별 링크들어가서 해당 주식 정보 파악
links<-GET(url=URL) %>%
  read_html() %>%
  html_nodes(css='td.ctg a')%>%
  html_attr('href')
a<-'https://finance.naver.com'
links<-str_c(a,links)
sise_links<-str_replace(links,'main','sise_day')
sise_links
rinklist<-list()
for (i in 1:10){
  res<-GET(url='https://finance.naver.com/item/sise_day.nhn?code=005930&',
           query=list(page=i))
  test<- res %>%
  read_html() %>%
  html_nodes(css='table.type2') %>%
  html_table()
rinklist<-append(rinklist, test)
}
rinklist
# do.call(rbind, rinlist) 다중리스트를 행으로(rbind) 다 붙이기
# list의 key별로 rbind로 한번에 붙여서 하나의 Df으로 만들기
do.call(rbind, rinklist)
Sys.setlocale(category = 'LC_ALL',locale='korean')

stock<-c()
for (i in 1:20){
  stock<-c(stock, read_html(paste0('https://finance.naver.com/sise/entryJongmok.nhn?&page=',i))%>%
             html_nodes(css='td.ctg a')%>%
             html_attr('href'))
}

stock<-str_c("https://finance.naver.com",stock)
stock<-str_replace(stock, 'main', 'sise_day')
stock
stock_total<-data.frame()
# url : 주식종목
for(url in stock){
  Sys.sleep(sample(1,1000,1))
  stock_sise<-data.frame()
  rslt<-list()
  for(i in 1:10){
    #test : 종목별 일별시세 다량 10page까지링크
    test<- read_html(paste0(url,"&page=",i)) %>%
      html_nodes(css='table.type2') %>%
      html_table()
    rslt<-append(rslt,test)
  }
  stock_sise<-do.call(rbind,rslt)
  # 무슨 주식종목인지 변수 추가해줌!
  stock_sise$name<-str_sub(url,nchar(url)-5,nchar(url))
    stock_total<-rbind(stock_total, stock_sise)
}
print(stock_total)

# 날짜가 공백이 아닌거만 담아
stock_total<-stock_total[stock_total$날짜!='',]
stock_total
nrow(stock_total)
head(stock_total)

## 주식간의 상관성 시각화##
# 변수타입확인
sapply(stock_total,class)
# 문자변수 모두 숫자형으로 바꾸어주기
numeric_fn<-function(x){
  x<-as.numeric(gsub(',','',x))
}
#날짜
final<-cbind(stock_total[1],
             as.data.frame(lapply(stock_total[2:7],numeric_fn)),
             stock_total[length(stock_total)]) 
head(final)
# 날짜칼럼 문자열 날짜로 인식하게 하기
library(lubridate)
final$날짜<-as.Date(final$날짜,format='%Y.%m.%d')
#날짜별로 정렬
final<-final %>%
  arrange(name, 날짜)
library(dplyr)
# name별로 종가가격 차이 구하기
final2<-final %>%
  group_by(name) %>%
  # . 은 앞의 데이터가져오라는뜻 여기서는 final 
  # sapply는 벡터로 반환
  do(as.data.frame(sapply(.[2:7], diff))) %>%
  # do함수로 rbind함수부르기
  rbind
final2
library(stringr)
stock_code<-str_sub(stock,nchar(stock)-5,nchar(stock))
# 네트워크를 만들기 위해 조합을 만들어야함
stock_comb<-combn(stock_code, 2)
t(stock_comb)
save(stock_total, file="C:/Users/joyh1/Desktop/빅데이터_20-1/naver_stock.Rdata")
load("C:/Users/joyh1/Desktop/빅데이터_20-1/naver_stock.Rdata")
stock_total
stock_comb<-t(stock_comb)
#5.4일부터
stock_c<-c()
for (i in 1:nrow(stock_comb)){
  a<-final2[final2$name==stock_comb[i,1],'종가']
  b<-final2[final2$name==stock_comb[i,2], '종가']
  print(i)
  # cor(a,b)인자에는 벡터 들어갈 수 있음
  stock_c<-c(stock_c,cor(a,b))
}
# 1번째,2번째 벡터 제외하고 할당(ppt자료근거)
stock_c<-stock_c[-1:-2]
result<-as.data.frame(stock_comb)
result<-cbind(result,stock_c)
colnames(result)<-c("Source","Target","weight")
head(result,n=5)
# cor값이 0.8이상이면강함, 0.4~0.8이면 moderate, 0.2이하면 약함
summary(result$weight)
save(result, file="C:/Users/joyh1/Desktop/빅데이터_20-1/result.Rdata")
load("C:/Users/joyh1/Desktop/빅데이터_20-1/result.Rdata")
result
```
