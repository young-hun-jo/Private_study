```r
library(dplyr)
library(lubridate)

# date 날짜형으로 인식하고 날짜와 비트코인 종류별로 정렬
data$date<-as.Date(data$date, format='%Y-%m-%d')
mid_data<-data %>% arrange(name, date)
# 전일대비 증감 변수 생성
final_data<-mid_data %>% group_by(name) %>% do(as.data.frame(sapply(.[3:6], diff))) %>% rbind()
final_data
```

```r
# 비트코인 종류 할당
bit_code<-unique(final_data$name)
# factor형을 vector로 변환
bit_code<-as.character(bit_code)
bit_code
```

```r
# 비트코인 종류별로 조합 생성
bit_comb<-combn(bit_code, 2)
bit_comb<-t(bit_comb)
head(bit_comb)
```

```r
bit_comb
```

```r
final_data
```

```r
# 비트코인 조합별로 'high price' 상관관계 저장할 벡터 생성
# 조합 한줄씩 읽어오기
# 비트코인 종류마다 모든 날짜의 가격이 있는게 아니기 때문에 비트코인 종류 조합(2개의 비트코인 종류)중 적은 데이터에 맞춰서 상관관계 생성

high_price_corr<-c()
for (i in 1:nrow(bit_comb)){
    a<-final_data[final_data$name == bit_comb[i,1], 'high_price']
    b<-final_data[final_data$name == bit_comb[i,2], 'high_price']
    if (length(pull(a)) > length(pull(b))){
        a<-pull(a)[1:length(pull(b))]
    } else if(length(pull(a)) < length(pull(b))){
        b<-pull(b)[1:length(pull(a))]
    }
    high_price_corr<-c(high_price_corr, cor(a,b))
}
# 조합개수와 상관관계 개수 일치하는지 확인(조합은 matrix로 되어있어서 row개수 출력으로 비교)
nrow(bit_comb)
length(high_price_corr)
```

```r
low_price_corr<-c()
for (i in 1:nrow(bit_comb)){
    a<-final_data[final_data$name == bit_comb[i,1], 'low_price']
    b<-final_data[final_data$name == bit_comb[i,2], 'low_price']
    if (length(pull(a)) > length(pull(b))){
        a<-pull(a)[1:length(pull(b))]
    } else if(length(pull(a)) < length(pull(b))){
        b<-pull(b)[1:length(pull(a))]
    }
    low_price_corr<-c(low_price_corr, cor(a,b))
}
# 조합개수와 상관관계 개수 일치하는지 확인(조합은 matrix로 되어있어서 row개수 출력으로 비교)
nrow(bit_comb)
length(low_price_corr)
```

```r
open_price_corr<-c()
for (i in 1:nrow(bit_comb)){
    a<-final_data[final_data$name == bit_comb[i,1], 'open_price']
    b<-final_data[final_data$name == bit_comb[i,2], 'open_price']
    if (length(pull(a)) > length(pull(b))){
        a<-pull(a)[1:length(pull(b))]
    } else if(length(pull(a)) < length(pull(b))){
        b<-pull(b)[1:length(pull(a))]
    }
    open_price_corr<-c(open_price_corr, cor(a,b))
}
# 조합개수와 상관관계 개수 일치하는지 확인(조합은 matrix로 되어있어서 row개수 출력으로 비교)
nrow(bit_comb)
length(open_price_corr)
```

```r
close_price_corr<-c()
for (i in 1:nrow(bit_comb)){
    a<-final_data[final_data$name == bit_comb[i,1], 'close_price']
    b<-final_data[final_data$name == bit_comb[i,2], 'close_price']
    if (length(pull(a)) > length(pull(b))){
        a<-pull(a)[1:length(pull(b))]
    } else if(length(pull(a)) < length(pull(b))){
        b<-pull(b)[1:length(pull(a))]
    }
    close_price_corr<-c(close_price_corr, cor(a,b))
}
# 조합개수와 상관관계 개수 일치하는지 확인(조합은 matrix로 되어있어서 row개수 출력으로 비교)
nrow(bit_comb)
length(close_price_corr)
```

```r
cor_result<-as.data.frame(bit_comb)
cor_result<-cbind(cor_result, high_price_corr, low_price_corr, open_price_corr, close_price_corr)
colnames(cor_result)<-c("Source","Target","high_Weight","low_Weight","open_Weight","close_Weight")
head(cor_result)
```

```r
summary(cor_result$high_Weight)
summary(cor_result$low_Weight)
summary(cor_result$open_Weight)
summary(cor_result$close_Weight)
```
